```python
import xarray as xr


url = "http://thredds.secoora.org/thredds/dodsC/SECOORA_NCSU_CNAPS.nc"
ds = xr.open_dataset(url)
```

```python
ds["temp"]
```

```python
surface = ds["temp"].sel(time="2019-07-03").isel(s_rho=35).squeeze()
```

```python
%matplotlib inline

import matplotlib.pyplot as plt


surface.plot.pcolormesh(x="lon_rho", y="lat_rho", figsize=(9, 9));
```
